## 5-Programs of Array Manipulation

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.array([[5,6]])
print("concate with axis=0:",np.concatenate((a,b),axis=0))
print("concate with axis=1:",np.concatenate((a,b.T),axis=1))

print(np.concatenate((a,b),axis=None))

a=np.array([[12,4,5],[23,45,66],[45,34,23]])
b=np.array([[1,40,50],[2,4,6],[4,3,2]])

#vertical stacking
print(np.vstack((a,b)))
              
#Horizontal stacking
print(np.hstack((a,b)))
              
print(a.reshape(3,3))

```

### 5-A1.searching

```python
import numpy as np
arr=np.array([1,2,3,4,5,4,4])
x=np.where(arr==4)
print(x)

arr=np.array([6,7,8,9])
x=np.searchsorted(arr,5)
print(x)

arr=np.array([1,3,5,7])
x=np.searchsorted(arr,[2,4,6])
print(x)
```

### 5-A2.Sorting

```python
#importing numpy packing
import numpy as np
a=np.array([[1,4],[3,1]])
print("Sorted array:",np.sort(a))
print("\n Sorted flattened array:",np.sort(a,axis=0))

x=np.array([3,1,2])
print("\n Indices that would sort an array",np.argsort(x))
print("\n Sorted complex number:",np.sort_complex([[3+4j,1-2j,5+1j,2+2j]]))
```

### 5-A3.Splitting

```python

import numpy as np
x=np.arange(9.0)
print(x)
print(np.split(x,3))
print(np.split(x,[3,5,6,10]))

x=np.arange(9)
print(np.array_split(x,4))

a=np.array([[1,3,5,7,9,11],[2,4,6,8,10,12]])
print("Splitting along horizontal axis into 2 parts:\n",np.hsplit(a,2))
print("\n Splitting along Vertical axis into 2 parts:\n",np.vsplit(a,2))
```

### 5-B2.Plotting

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,3*np.pi,0.1)
print("x=",x)

y_sin=np.sin(x)
y_cos=np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('x value')
plt.ylabel('y sine and cosine values')
plt.title('Sine and Cosine')
plt.legend(['Sine','Cosine'])

plt.show()
```
